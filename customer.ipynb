```python
# Load and prepare data for database
import pandas as pd

df = pd.read_csv('customer_data.csv')

# Clean column names
df.columns = df.columns.str.strip()

# Fill missing Review Rating by Category median
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))

# Create Age groups
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['Age_group'] = pd.qcut(df['Age'], q=4, labels=labels)

# Map purchase frequency to days
frequency_mapping = {
    'fortnightly': 14,
    'monthly': 30,
    'weekly': 7,
    'quarterly': 90,
    'bi-weekly': 14,
    'Annually': 365,
    'Every 3 months': 90
}
df['purchase_frequency_days'] = df['Frequency of Purchases'].map(frequency_mapping).fillna(0)

# Drop duplicate column
df = df.drop('Promo Code Used', axis=1)

df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'Purchase_Amount_(USD)': 'Purchase_Amount'})


print(f"Data loaded and prepared. Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
```

```python
from sqlalchemy import create_engine

#Mysql connection 
username = 'root'
password = 'password'
host = 'localhost'
database = 'customer_behavior'    

engine = create_engine(f'mysql+pymysql://{username}:{password}@{host}/{database}')

table_name = "customer_behavior"

# ensure the database exists before writing the table
server_engine = create_engine(f'mysql+pymysql://{username}:{password}@{host}/')
with server_engine.connect() as conn:
    conn.exec_driver_sql(f"CREATE DATABASE IF NOT EXISTS `{database}`")

engine = create_engine(f'mysql+pymysql://{username}:{password}@{host}/{database}')
df.to_sql(table_name, engine, if_exists='replace', index=False)

pd.read_sql("SELECT * FROM customer_behavior LIMIT 10;", engine)

```
